```python
from TTS.api import TTS

# Load XTTS v2 Model (Uses GPU if available)
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2", gpu=True)

# Define input text and reference voice file
text = "It took me quite a long time to develop a voice, and now that I have it, I'm not going to be silent."
reference_audio = "reference.wav"  # Change this to the actual path of your reference voice file
output_path = "output.wav"
```

```python
import IPython.display as ipd
from TTS.api import TTS
import numpy as np
import soundfile as sf
import io
def generate_audio(text, reference_audio):
    """Generate speech and play it in Jupyter Notebook."""
    # Save audio to a buffer instead of a file
    buffer = io.BytesIO()

    # Generate speech
    tts.tts_to_file(
        text=text,
        file_path="temp.wav",  # Temporary file
        speaker_wav=reference_audio,  # Reference voice file
        language="en"
    )

    # Load the file into memory
    data, samplerate = sf.read("temp.wav")
    buffer.seek(0)

    # Convert to NumPy array
    audio_array = np.array(data, dtype=np.float32)

    # Play the audio in Jupyter
    return ipd.Audio(audio_array, rate=samplerate)
```

```python
# Define multiple reference voice files
reference_audio_files = ["ref_1.wav", "ref_1.wav", "ref_3.wav"]  # List of reference audios

text_to_speak = """I'm happy to help you with any questions or concerns you may have. Is there something specific you'd like to know or discuss about our services?"""
generate_audio(text_to_speak, reference_audio_files)
```
